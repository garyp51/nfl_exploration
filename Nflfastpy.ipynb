```python
import nflfastpy
import pandas as pd
import matplotlib as plt
%matplotlib inline
```

```python
'''Read in a single year of data'''
df = nflfastpy.load_pbp_data(year=2021)
```

```python
'''Read in multiple years, set range to the years you want and it will create a full dataframe'''

years = range(2019, 2022)
df = []
for year in years:
    df_year = nflfastpy.load_pbp_data(year=year)
    df.append(df_year)
df = pd.concat(df)
df
```

```python
''' read in 49er games only'''

sf = df[(df['home_team'] == 'SF') | (df['away_team'] == 'SF')]
```

```python
sf
```

```python
pass_cols = ['home_team', 'away_team', 'week','posteam', 'passer', 'passer_player_name', 'passing_yards','pass','name','pass_length','pass_location','air_epa','yac_epa','first_down_pass','incomplete_pass','interception','pass_touchdown','complete_pass','cp','cpoe','xpass','pass_oe','qb_epa']
```

```python
sf_pass = sf[pass_cols]
sf_pass = sf_pass[(sf_pass['pass'] == 1) & (sf_pass['posteam'] == 'SF')]
```

```python
sf_pass
```

```python
sf_pass.groupby(['passer']).sum()
```

```python
sf_pass.groupby(['passer']).mean()
```

```python
sf_pass.groupby(['passer']).mean().plot.bar()
```
